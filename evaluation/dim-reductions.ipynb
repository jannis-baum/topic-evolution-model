```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE

import matplotlib.pyplot as plt

from helpers import import_data
```

```python
X, y = import_data()
```

```python
sc = StandardScaler()
X = sc.fit_transform(X)
labels = pd.Categorical(y)
```

```python
pca = PCA()
X_pca = pca.fit_transform(X)
pca.explained_variance_ratio_

print(f'Explained variance is {sum(pca.explained_variance_ratio_[:2]) * 100 :.0f}%')
plot = plt.scatter(X_pca[:,0], X_pca[:,1], c=labels.codes, alpha=0.3)
plt.legend(handles=plot.legend_elements()[0], labels=list(labels.categories))
plt.show()
```

```python
tsne = TSNE(2)
X_tsne = tsne.fit_transform(X)
```

```python
plot = plt.scatter(X_tsne[:,0], X_tsne[:,1], c=labels.codes, cmap='tab20', alpha=0.3)
plt.legend(handles=plot.legend_elements()[0], labels=list(labels.categories))
plt.show()
```
